```python
import pandas as pd
import numpy as np
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
path='https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/datasets_228_482_diabetes.csv'
data=pd.read_csv(path)
```

```python
data.info()
```

```python
data
```

```python
X=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
model=XGBClassifier(booster='gbtree',learning_rate=0.1,n_estimators=100,max_leaves=100,objective='reg:logistic')
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
params={
    'learning_rate':[0.01,0.1,1.8],
    'max_depth':[10,20,30],
    'multisample':[0.8,0.1,0.1],
    'n_estimator':[100,200,300]
}
```

```python
grid_search=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=5)
```

```python
grid_search.fit(X_train,y_train)
print(grid_search.best_params_)
y_pred=grid_search.predict(X_test)
print(accuracy_score(y_test,y_pred))
```
