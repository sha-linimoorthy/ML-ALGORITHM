**Support Vector Machine**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn import svm
import numpy as np
```

```python
test_data=pd.read_csv("https://raw.githubusercontent.com/nunnarilabs/ml/master/titanic/test.csv")
train_data=pd.read_csv("https://raw.githubusercontent.com/nunnarilabs/ml/master/titanic/train.csv")
```

```python
train_data
```

```python
train_data.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
train_data
```

```python
train_data[train_data.Embarked.isnull()]
```

```python
train_data.Embarked.value_counts()
```

```python
train_data.drop(['Cabin'],axis=1,inplace=True)
train_data['Age']=train_data['Age'].fillna(train_data['Age'].median())
train_data['Embarked']=train_data['Embarked'].fillna('C')
train_data.Sex=train_data.Sex.replace(['male','female'],[0,1])
train_data.Embarked=train_data.Embarked.replace(['S','C','Q'],[0,1,2])
```

```python
train_data.dropna(inplace=True)
```

```python
test_data.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
test_data.drop(['Cabin'],axis=1,inplace=True)
test_data.Sex=train_data.Sex.replace(['male','female'],[0,1])
test_data.Age=test_data.Age.fillna(test_data.Age.median())
test_data.Embarked=test_data.Embarked.fillna('C')
test_data.Embarked=test_data.Embarked.replace(['S','C','Q'],[0,1,2])
```

```python
test_data.dropna(inplace=True)
```

```python
features=['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']
```

```python
C=0.7 #margin
clf=svm.SVC(kernel='linear',C=C).fit(train_data[features],train_data['Survived'])
```

```python
y_pred=clf.predict(test_data)
```
